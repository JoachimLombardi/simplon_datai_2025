```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 50, 100)
```

```python
def cercles_concentriques(pas):
    for n in range (1, 5):
        plt.scatter(np.cos(x)/pas,np.sin(x)/pas)
        pas += pas/np.exp(n/2)
    plt.axis("scaled")
```

```python
cercles_concentriques(0.1)
```

```python

plt.plot(np.cos(x) + np.sin(20*x)/40, np.sin(x) + np.cos(20*x)/40)
plt.axis("scaled")
```
