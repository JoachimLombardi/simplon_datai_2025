```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(data_covid_sample.jour[::40], data_covid_sample["rea.1"][::40], label="réanimation-hommes")
ax.plot(data_covid_sample.jour[::40], data_covid_sample["rea.2"][::40], label="réanimation-femmes")
ax.plot(data_covid_sample.jour[::40], data_covid_sample["hosp.1"][::40], label="hospitalisation-hommes")
ax.plot(data_covid_sample.jour[::40], data_covid_sample["hosp.2"][::40], label="hospitalisation-femmes")
plt.xticks(rotation=45)
ax.legend()
```

```python
fig ,ax = plt.subplots(figsize=(10, 7))
ax.plot(data_covid_sample.jour[::40], data_covid_sample["dc.1"][::40], label="décès_cum_hommes")
y1, y2 = ax.get_ylim() 
ax.twinx().set_ylim(y1*10, y2*10)
ax.plot(data_covid_sample.jour[::40], data_covid_sample["dc.2"][::40], label="décès_cum_femmes")
y1, y2 = ax.get_ylim() 
ax.twinx().set_ylim(y1*10, y2*10)
ax.legend()
```

```python
nb_hommes = 32920000
nb_femmes = 35122000
debut_mort_homme = data_covid_sample.iloc[0]["dc.1"]
debut_mort_femme = data_covid_sample.iloc[0]["dc.2"]
fin_mort_homme = data_covid_sample.iloc[-1]["dc.1"] - data_covid_sample.iloc[-2]["dc.1"]
fin_mort_femme = data_covid_sample.iloc[-1]["dc.2"] - data_covid_sample.iloc[-2]["dc.2"]
```

```python
prob_mort_homme_debut = f"{debut_mort_homme/nb_hommes * 100:.5f} %" 
prob_mort_femme_debut = f"{debut_mort_femme/nb_femmes * 100:.5f} %"
prob_mort_homme_fin = f"{fin_mort_homme/nb_hommes * 100:.5f} %"
prob_mort_femme_fin = f"{fin_mort_femme/nb_femmes * 100:.5f} %" 
```

```python
mortalite = {}
mortalite["mortalite_homme_au_debut"] = prob_mort_homme_debut
mortalite["mortalite_femme_au_debut"] = prob_mort_femme_debut
mortalite["mortalite_homme_a_la_fin"] = prob_mort_homme_fin
mortalite["mortalite_femme_a_la_fin"] = prob_mort_femme_fin
mortalite
```
