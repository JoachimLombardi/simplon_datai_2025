```python
def affine(a,b,x1,x2):
  y = list()
  xx = list()
  x = range(x1,x2)
  for n in x :
    y.append(a*n+b)
    xx.append(n)
  return np.array(y), xx

a = .1
b =  0
x1 = -10
x2 = 10
y,x = affine(a,b,x1,x2)

import matplotlib.pyplot as plt

plt.plot(x,y)
plt.grid()
```

```python
nn = y.shape[0]
nn
```

```python
yerr = y + np.random.randn(nn)/10 
```

```python
plt.plot(x,y)
plt.plot(x,yerr,'o')
plt.grid()
```

```python
y_gauss = np.random.randn(nn)/10 # Representtion gaussienne
plt.plot(x, y_gauss, "o")
```

```python
def erreur(verite,prediction):
  err = np.mean( (verite-prediction)**2  )
  return err


list_err = list()
aaaaaa = np.arange(-2,2,.1)
for a in list(aaaaaa):
  y_test,x = affine(a,0,x1,x2)
  list_err.append(erreur(y,y_test))

plt.plot(aaaaaa,list_err)
plt.grid()
```

```python
list_err = np.array(list_err)
type(list_err)
a_calculer = aaaaaa[list_err.argmin()]
```

```python
# Courbe en cloche
def parabole(a, b, c, x0, x1):
    y = []
    x = np.array(range(x0, x1))
    for n in x:
        y.append(a*n**2 + b*n + c)
    return np.array(y), x
```

```python
a = 3
b = 2
c =1
x0 = -15
x1 = 15
```

```python
y, x = parabole(a, b, c, x0, x1)
plt.plot(x, y)
```

```python
nn = y.shape[0]
nn
```

```python
yerr = y + np.random.randn(nn)*50
```

```python
plt.plot(x, y)
plt.plot(x, yerr, "o")
```

```python
list_err = []
aa = np.array(range(2,4))
for a in aa:
    y_test, x = parabole(a, b=2, c=1, x0=-15, x1=15)
    list_err.append(erreur(y, y_test))
plt.plot(aa, list_err)

```
