```python
import matplotlib.pyplot as plt
```

```python
# On peut ne rentrer que les valeurs de y, x partira de 0 avec un pas de 1
plt.plot([1, 0, 3, 2]) # ici x va donc valoir 0, 1, 2, 3
plt.grid()
```

```python
# Mais on peut aussi choisir des valeures pour x
x = [-3, -1, 2]
y = [4, -4, -3]
plt.plot(x,y)
plt.grid()
```

```python
# Nos deux courbes peuvent galement apparatre sur un mme graphique
plt.plot([1, 0, 3, 2])
plt.plot(x,y)
plt.grid()
```

```python
# Tracer un polynme de degr 2 avec numpy et dcorer le tout
import numpy as np
x = np.linspace(-50, 50, 40)
fig, ax = plt.subplots(figsize=(7, 5))
ax.plot(x, x**2, "-+")
ax.set(title="Parabole",
       xlabel="axe x",
       ylabel="axe y")
ax.grid()
```

```python
# On a stock la figure dans un objet on peut donc la rcuprer facilement
fig
```

```python
# On peut galement diviser le graphe en plusieurs sous-graphes
fig, (ax0, ax1) = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(7, 7),
                            sharex=True)
ax0.plot(x, 1/x)
ax0.set(title="Hyperbole",
        ylabel="axe y")
ax0.set_ylim([-0.5, 0.25])

ax1.plot(x, 3*x)
ax1.set(title="Droite",
        xlabel="axe x",
        ylabel="axe y")
fig.suptitle(t="Affichage de deux figures");
```

```python
# On peut aussi le faire comme a
fig, axs = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(7, 7),
                            sharex=True)
axs[0].plot(x, 1/x)
axs[0].set(title="Hyperbole",
        ylabel="axe y")
axs[0].set_ylim([-0.5, 0.25])

axs[1].plot(x, 3*x)
axs[1].set(title="Droite",
        xlabel="axe x",
        ylabel="axe y")

fig.suptitle(t="Affichage de deux figures");
```

```python
# Ajoutons un camembert et un nuage de points
categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]

fig_total, axs = plt.subplots(nrows=2,
                              ncols=2,
                              figsize=(12, 12))
                        
axs[0, 0].plot(x, 1/x)
axs[0, 0].set(title="Hyperbole",
              xlabel="axe x",
              ylabel="axe y")

axs[0, 1].plot(x, x**2, "-+", label="Parabole")
axs[0, 1].set(title="Parabole",
              xlabel="axe x",
              ylabel="axe y")
axs[0, 1].plot(x, 20*x + 1000, 
               "-+", 
               label="Droite")
axs[0, 1].legend()

axs[1, 0].pie(valeurs, 
              labels=categories, 
              explode = [0, 0.3, 0, 0], 
              colors=["gold", "yellow", "purple", "indigo"],
              shadow=True,
              labeldistance=0.8,
              autopct = lambda x: str(round(x, 2)) + '%',
              pctdistance = 0.4)
axs[1, 0].set(title="Camembert")

axs[1, 1].scatter(np.linspace(-200, 200, 500), np.random.randn(500))
axs[1, 1].axhline(np.random.randn(500).mean(),
                  linestyle='--',
                  c="black")
axs[1, 1].set(title="Nuage de points",
              xlabel="axe x",
              ylabel="axe y")

fig_total.suptitle(t="Affichage de quatre figures", fontsize = 16, fontweight = "bold");
```

```python
# Une autre faon de faire avec les tuples
categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]

fig_total, ((ax0, ax1), (ax2, ax3)) = plt.subplots(nrows=2,
                                                   ncols=2,
                                                   figsize=(12, 12))
                                            
ax0.plot(x, 1/x)
ax0.set(title="Hyperbole",
        xlabel="axe x",
        ylabel="axe y")

ax1.plot(x, x**2, "-+", label="Parabole")
ax1.set(title="Parabole",
        xlabel="axe x",
        ylabel="axe y")
ax1.plot(x, 20*x + 1000,
         "-+", 
         label="Droite")
ax1.legend()

ax2.pie(valeurs, 
        labels=categories, 
        explode = [0, 0.3, 0, 0], 
        colors=["gold", "yellow", "purple", "indigo"],
        shadow=True,
        labeldistance=0.8,
        autopct = lambda x: str(round(x, 2)) + '%',
        pctdistance = 0.4)
ax2.set(title="Camembert")

ax3.scatter(np.linspace(-200, 200, 500), np.random.randn(500))
ax3.axhline(np.random.randn(500).mean(),
            linestyle='--', 
            c="black")
ax3.set(title="Nuage de points",
        xlabel="axe x",
        ylabel="axe y")

fig_total.suptitle(t="Affichage de quatre figures", fontsize = 16, fontweight = "bold");
```
